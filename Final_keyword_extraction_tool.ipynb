```python
def build_vocabulary(page:list) -> list:
    vocab = list(set(page))
    vocab.sort()
    
    vocab_dict = {}
    for index, word in enumerate(vocab):
        vocab_dict[word] = index
    return vocab_dict
vocab_dict = build_vocabulary(text2)
print(vocab_dict)
```
